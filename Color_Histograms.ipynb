```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


# Read in the image

def plot_histogram(image):
    # Take histograms in R, G, and B
    rhist = np.histogram(image[:,:,0], bins=32, range=(0, 256))
    ghist = np.histogram(image[:,:,1], bins=32, range=(0, 256))
    bhist = np.histogram(image[:,:,2], bins=32, range=(0, 256))

    # Generating bin centers
    bin_edges = rhist[1]
    bin_centers = (bin_edges[1:]  + bin_edges[0:len(bin_edges)-1])/2
    plt.imshow(image)
    fig = plt.figure(figsize=(12,3))
    plt.subplot(131)
    plt.bar(bin_centers, rhist[0])
    plt.xlim(0, 256)
    plt.title('R Histogram')
    plt.subplot(132)
    plt.bar(bin_centers, ghist[0])
    plt.xlim(0, 256)
    plt.title('G Histogram')
    plt.subplot(133)
    plt.bar(bin_centers, bhist[0])
    plt.xlim(0, 256)
    plt.title('B Histogram')
    plt.show()

image = mpimg.imread('../Images/explore_color_spaces/vehicle1.jpeg')
plot_histogram(image)
image = mpimg.imread('../Images/explore_color_spaces/vehicle2.jpeg')
plot_histogram(image)
image = mpimg.imread('../Images/explore_color_spaces/vehicle3.jpeg')
plot_histogram(image)

```

```python
image = mpimg.imread('../Images/explore_color_spaces/nonvehicle1.jpeg')
plot_histogram(image)
image = mpimg.imread('../Images/explore_color_spaces/nonvehicle2.jpeg')
plot_histogram(image)
image = mpimg.imread('../Images/explore_color_spaces/nonvehicle3.jpeg')
plot_histogram(image)
```
